# Individual Model Evaluation 

This notebook details the performance of the following algorithmic approaches:

- DQN
- Double DQN
- Double Dueling DQN
- DDPG
- DDQN-PER

The end of the notebook display a graph of all agent training performances.


## DQN

```python
import pickle
import matplotlib.pyplot as plt

with open("./dqn/train_rewards_list.p", "rb") as fp: 
    offline_dqn_rewards = pickle.load(fp)
    
with open("./dqn/train_rewards_avg_list.p", "rb") as fp:
    offline_dqn_avg_rewards = pickle.load(fp)    
```

```python
plt.title("Learning Curve")
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.plot(offline_dqn_rewards)

plt.xlabel("Episode")
plt.ylabel("Reward")
plt.plot(offline_dqn_avg_rewards, 'r')
```

## Double DQN

```python
import pickle
import matplotlib.pyplot as plt

with open("./double-dqn/double-dqn-rewards", "rb") as fp: 
    offline_rewards = pickle.load(fp)
    
with open("./double-dqn/double-dqn-avg-rewards", "rb") as fp:
    offline_avg_rewards = pickle.load(fp)    
```

# DDPG

```python
import pandas as pd
data = pd.read_csv('./ddpg/logs/ddpg_progress.log', sep=';')
data.head()
plt.plot(data['reward'])
plt.plot(data['average'])
plt.title('Reward')
plt.ylabel('Reward')
plt.xlabel('Episode')
plt.legend(['Reward'], loc='best')
plt.show()
```

# DDQN-PER

```python
import numpy as np
offline_avg_ddqn_per= np.load("ddqn-per/results/results_DDQN_PER_avg_1650732510.npy")
offline_sum_ddqn_per= np.load("ddqn-per/results/results_DDQN_PER_sum_1650732510.npy")
plt.plot(offline_sum_ddqn_per)
plt.plot(offline_avg_ddqn_per, color="red")
```

## Overall Comparison

```python
plt.figure(figsize=(15, 8))

plt.title("Learning Curve")
plt.xlabel("Episode")
plt.ylabel("Average Reward")


plt.plot(offline_dqn_avg_rewards, label = "DQN") # dqn
plt.plot(offline_avg_rewards, label = "Double DQN") # ddqn
#plt.plot(offline_dddqn_avg_rewards, label = "DDDQN") # dddqn
plt.plot(data['average'], label = "DDPG") # DDPG
plt.plot(offline_avg_ddqn_per, label = "DDQN + PER") # DDPG
plt.legend()
```
